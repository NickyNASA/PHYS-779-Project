```python
import numpy as np
from numpy import pi, cos, sin

import scipy as sp
from scipy.integrate import simpson

import matplotlib.pyplot as plt


def I_cos_DRAG(t_arr, tp):
    return 1 - np.cos(2*pi*t_arr/tp)

def Q_cos_DRAG(t_arr, tp, alpha, beta):
    # -beta/alpha * d/dt(omega_I)
    return -(beta/alpha) * (2*pi/tp) * np.sin(2*pi*t_arr/tp)
```

Constants

```python
anharmonicity = -0.212 # GHz
alpha = 2*pi*anharmonicity

tp = 6 # ns
n = 50
t_arr = np.linspace(0, tp, n)
```

Cosine DRAG

```python
beta = 1 # DRAG-L
theta = pi # Desired rotation angle

I_arr = I_cos_DRAG(t_arr, tp)
amp = theta / simpson(I_arr, x=t_arr)
I_arr = amp * I_arr

Q_arr = amp * Q_cos_DRAG(t_arr, tp, alpha, beta)

fig, ax = plt.subplots(1, 1)
plt.plot(t_arr, I_arr)
plt.plot(t_arr, Q_arr)
plt.title("Cos DRAG")
plt.xlim(0, tp)
plt.ylim(-1.2, 1.2)
plt.show()
```
