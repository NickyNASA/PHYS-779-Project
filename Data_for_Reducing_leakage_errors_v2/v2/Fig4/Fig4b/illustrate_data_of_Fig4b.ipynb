```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
#Re-read csv-files and plot
dark_blue = [51/255,51/255,255/255] #[65,105,225]/255; %royal blue
crimson = [220/255, 20/255, 60/255] 
light_red = [240/255,128/255,128/255]
dark_gray = [80/255, 80/255, 80/255] #[64/255, 64/255, 64/255]
colors_list = [dark_blue, dark_gray, crimson]        


gate_error_vs_time = np.genfromtxt('gate_error_vs_time.csv', delimiter=',')
gate_error_uncertainty_vs_time = np.genfromtxt('gate_error_uncertainty_vs_time.csv', delimiter=',')
incoherent_error_vs_time = np.genfromtxt('incoherent_error_vs_time.csv', delimiter=',')
incoherent_error_uncertainty_vs_time = np.genfromtxt('incoherent_error_uncertainty_vs_time.csv', delimiter=',')
leakage_error_vs_time = np.genfromtxt('leakage_error_vs_time.csv', delimiter=',')
leakage_error_uncertainty_vs_time = np.genfromtxt('leakage_error_uncertainty_vs_time.csv', delimiter=',')
coherence_limit_estimate_vs_time = np.genfromtxt('coherence_limit_estimate_vs_time.csv', delimiter=',')
time_arr = np.genfromtxt('time_arr.csv', delimiter=',')


fig, ax = plt.subplots(1, 1, figsize=(5,4))
plt.errorbar(time_arr, gate_error_vs_time, yerr=gate_error_uncertainty_vs_time, fmt='^', color=colors_list[0]) #label = 'Total error')
plt.text(2, 2e-4, 'RB error', fontsize = 14.5, color=colors_list[0]) 
plt.errorbar(time_arr, incoherent_error_vs_time, yerr=incoherent_error_uncertainty_vs_time, fmt='o', color=colors_list[1]) #label = 'Incoherent error')
plt.text(2, 0.57e-4, 'Purity RB error', fontsize = 14.5, color=colors_list[1]) 
plt.errorbar(time_arr, leakage_error_vs_time, yerr=leakage_error_uncertainty_vs_time, fmt='s', color=colors_list[2]) #label = 'Leakage error')
plt.text(2, 3.4e-5, 'Leakage error', fontsize = 14.5, color=colors_list[2]) 

plt.plot(time_arr, coherence_limit_estimate_vs_time, '--', color=colors_list[1], label = 'Limit from ' + r'$T_1$ and $T_2^\mathrm{e}$')  
plt.xlabel('Time from calibration (h)', fontsize = 16)
#plt.ylabel(r'Gate error ($\times 10^{-4}$)', fontsize = 16)
plt.ylabel(r'Magnitude of error', fontsize = 16)
plt.title(r'FAST DRAG-L: $t_\mathrm{g}=6.67$ ns', fontsize = 15)

plt.legend(fontsize = 14, loc = 'upper right', frameon=False) #'upper right')
#plt.ylim([1e-4, 5e-4])
#plt.ylim([1, 6.5])
plt.ylim([1e-5, 4e-4])
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
ax.tick_params(length=8, width = 1.4)
ax.tick_params('both', length=4, width=1, which='minor')
for side in ['top', 'left', 'right', 'bottom']:
    ax.spines[side].set_linewidth(1.4)
ax.set_yscale("log")
```

```python
#T1 and T2 echo vs time
T1_mean_vs_time = np.genfromtxt('T1_mean_vs_time.csv', delimiter=',')
T1_uncertainty_vs_time = np.genfromtxt('T1_uncertainty_vs_time.csv', delimiter=',')
T2Echo_mean_vs_time = np.genfromtxt('T2Echo_mean_vs_time.csv', delimiter=',')
T2Echo_uncertainty_vs_time = np.genfromtxt('T2Echo_uncertainty_vs_time.csv', delimiter=',')


fig, ax = plt.subplots(1, 1, figsize=(5,4))
plt.errorbar(time_arr, 1e6*T1_mean_vs_time, yerr=1e6*T1_uncertainty_vs_time, fmt='^', color=colors_list[0], label = r'$T_1$')
plt.errorbar(time_arr, 1e6*T2Echo_mean_vs_time, yerr=1e6*T2Echo_uncertainty_vs_time, fmt='o', color=colors_list[1], label = r'$T_2^\mathrm{e}$') #label = 'Total error')
plt.xlabel('Time from calibration (h)', fontsize = 16)
#plt.ylabel(r'Gate error ($\times 10^{-4}$)', fontsize = 16)
plt.ylabel(r'$T_1$, $T_2^\mathrm{e}$ ($\mu$s)', fontsize = 16)
plt.title(r'FAST DRAG-L: $t_\mathrm{g}=6.67$ ns', fontsize = 15)

plt.legend(fontsize = 14, loc = 'upper right') #'upper right')
#plt.ylim([1e-4, 5e-4])
#plt.ylim([1, 6.5])
plt.ylim([0, 60])
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
ax.tick_params(length=8, width = 1.4)
ax.tick_params('both', length=4, width=1, which='minor')
for side in ['top', 'left', 'right', 'bottom']:
    ax.spines[side].set_linewidth(1.4)
```
