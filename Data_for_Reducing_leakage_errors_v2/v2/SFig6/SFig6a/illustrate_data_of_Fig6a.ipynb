```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import os
```

```python
#Re-read csv-files and plot
dark_blue = [51/255,51/255,255/255] #[65,105,225]/255; %royal blue
crimson = [220/255, 20/255, 60/255] 
dark_orange = [255/255, 128/255, 0]
dark_gray = [64/255, 64/255, 64/255]

color_list = [dark_blue, crimson, dark_orange, dark_gray]



fig, ax = plt.subplots(1, 1, figsize=(5.2,4))
indices_for_fig = [0, 4, 10, 16]
for j, idx in enumerate(indices_for_fig):
    #Experiments
    axis_shift_vs_delay_exp_arr = np.genfromtxt(os.path.join('idx_of_duration_%d'%(idx), 'axis_shift_vs_delay_exp_arr.csv'), delimiter=',')
    axis_shift_vs_delay_fit_arr = np.genfromtxt(os.path.join('idx_of_duration_%d'%(idx), 'axis_shift_vs_delay_fit_arr.csv'), delimiter=',')
    delay_exp_arr = np.genfromtxt(os.path.join('idx_of_duration_%d'%(idx), 'delay_exp_arr.csv'), delimiter=',')
    delay_fit_arr = np.genfromtxt(os.path.join('idx_of_duration_%d'%(idx), 'delay_fit_arr.csv'), delimiter=',')
    pulse_duration = np.genfromtxt(os.path.join('idx_of_duration_%d'%(idx), 'pulse_duration.csv'), delimiter=',')


    plt.plot(delay_exp_arr*1e9, axis_shift_vs_delay_exp_arr, 'o', color=color_list[j], label=r'$t_\mathrm{p}=%.1f$ ns'%(1e9*pulse_duration)) #, label = '%s'%(pulse_leg))
    plt.plot(delay_fit_arr*1e9, axis_shift_vs_delay_fit_arr, '-', color=color_list[j]) #, label = '%s'%(pulse_leg))

plt.xlabel(r'Delay between pulses, $t_\mathrm{d}$ (ns)', fontsize = 16)
plt.ylabel(r'Axis shift, $\varphi_\mathrm{s}$ (deg.)', fontsize = 16)
plt.legend(fontsize = 12, loc = 'lower right') #'upper right')
plt.title('CosRiseFall DRAG-P', fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
ax.tick_params(length=6, width = 1.4)
for side in ['top', 'left', 'right', 'bottom']:
    ax.spines[side].set_linewidth(1.4)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5.2,4))
for idx in range(18):
    #Experiments
    axis_shift_vs_delay_exp_arr = np.genfromtxt(os.path.join('idx_of_duration_%d'%(idx), 'axis_shift_vs_delay_exp_arr.csv'), delimiter=',')
    axis_shift_vs_delay_fit_arr = np.genfromtxt(os.path.join('idx_of_duration_%d'%(idx), 'axis_shift_vs_delay_fit_arr.csv'), delimiter=',')
    delay_exp_arr = np.genfromtxt(os.path.join('idx_of_duration_%d'%(idx), 'delay_exp_arr.csv'), delimiter=',')
    delay_fit_arr = np.genfromtxt(os.path.join('idx_of_duration_%d'%(idx), 'delay_fit_arr.csv'), delimiter=',')
    pulse_duration = np.genfromtxt(os.path.join('idx_of_duration_%d'%(idx), 'pulse_duration.csv'), delimiter=',')


    plt.plot(delay_exp_arr*1e9, axis_shift_vs_delay_exp_arr, 'o', color=np.array(dark_blue)*idx/17 + np.array(crimson)*(17-idx)/17, label=r'$t_\mathrm{p}=%.1f$ ns'%(1e9*pulse_duration)) #, label = '%s'%(pulse_leg))
    plt.plot(delay_fit_arr*1e9, axis_shift_vs_delay_fit_arr, '-', color=np.array(dark_blue)*idx/17 + np.array(crimson)*(17-idx)/17) #, label = '%s'%(pulse_leg))

plt.xlabel(r'Delay between pulses, $t_\mathrm{d}$ (ns)', fontsize = 16)
plt.ylabel(r'Axis shift, $\varphi_\mathrm{s}$ (deg.)', fontsize = 16)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.75, box.height])
plt.legend(fontsize = 12, loc = 'center left', bbox_to_anchor=(1, 0.5)) #'upper right')
plt.title('CosRiseFall DRAG-P', fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
ax.tick_params(length=6, width = 1.4)
for side in ['top', 'left', 'right', 'bottom']:
    ax.spines[side].set_linewidth(1.4)
```
