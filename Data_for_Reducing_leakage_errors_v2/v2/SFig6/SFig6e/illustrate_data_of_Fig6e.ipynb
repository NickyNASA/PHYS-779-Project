```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import os
```

```python
#Re-read csv-files and plot
light_blue = [0.3010, 0.7450, 0.9330]
dark_blue = [51/255,51/255,255/255] #[65,105,225]/255; %royal blue

cost_function_vs_a1_and_tp_mat = np.genfromtxt('cost_function_vs_a1_and_tp_mat.csv', delimiter=',')
pulse_duration_arr = np.genfromtxt('pulse_duration_arr.csv', delimiter=',')
a1_arr = np.genfromtxt('a1_arr.csv', delimiter=',')


color_list = [list(j / 3 * np.array(dark_blue) + (3 - j)/ 3 * np.array(light_blue)) for j in range(4)]
fig, ax = plt.subplots(1, 1, figsize=(5,4))
for j in range(len(pulse_duration_arr)):
    cost_fun_1d_arr = cost_function_vs_a1_and_tp_mat[j, :]
    plt.plot(a1_arr, cost_fun_1d_arr, 'o--', color=color_list[j], label=r'$t_\mathrm{p}=%.1f$ ns'%(1e9*pulse_duration_arr[j])) #, label = '%s'%(pulse_leg))
plt.plot([-0.028, -0.028], np.array([1e-2, 1e1]), '--', color='k', label=r'$a_1=-0.028$' + '\n for pre-distortion') 

plt.xlabel(r'$a_1$-coefficient', fontsize = 16)
plt.ylabel(r'Cost function, $\sum_i \varphi_\mathrm{s}(t_{\mathrm{d}, i})^2$', fontsize = 16)
plt.legend(fontsize = 11, loc = 'upper right') #'upper right')
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.title('FAST DRAG-P', fontsize=15)
ax.tick_params(length=6, width = 1.4)
for side in ['top', 'left', 'right', 'bottom']:
    ax.spines[side].set_linewidth(1.4)
ax.set_yscale("log")
plt.ylim([2e-2, 1e1])
```
