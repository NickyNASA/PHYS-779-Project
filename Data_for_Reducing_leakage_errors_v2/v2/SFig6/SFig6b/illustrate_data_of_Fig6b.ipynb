```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import os
```

```python
#Re-read csv-files and plot
light_blue = [0.3010, 0.7450, 0.9330]

dark_blue = [51/255,51/255,255/255] #[65,105,225]/255; %royal blue
crimson = [220/255, 20/255, 60/255] 
dark_orange = [255/255, 128/255, 0]
dark_gray = [64/255, 64/255, 64/255]

color_list = [dark_blue, crimson, dark_orange, dark_gray]



decay_time_of_axis_shift_vs_pulse_duration_arr = np.genfromtxt('decay_time_of_axis_shift_vs_pulse_duration_arr.csv', delimiter=',')
pulse_duration_arr = np.genfromtxt('pulse_duration_arr.csv', delimiter=',')
#indices corresponding to traces in panel (a)
duration_indices = [0, 4, 10, 16]

fig, ax = plt.subplots(1, 1, figsize=(5,4))
plt.plot(pulse_duration_arr*1e9, 1e9*decay_time_of_axis_shift_vs_pulse_duration_arr, 'o', color=light_blue, label='Experiment') 
for j, duration_idx in enumerate(duration_indices):
    plt.plot(pulse_duration_arr[duration_idx]*1e9, 1e9*decay_time_of_axis_shift_vs_pulse_duration_arr[duration_idx], 'o', markerfacecolor="None", markeredgecolor=color_list[j], markersize = 13, markeredgewidth= 2) #, label = '%s'%(pulse_leg))
plt.plot(pulse_duration_arr*1e9, 8*np.ones(pulse_duration_arr.shape), '--k', label=r'$\tau_1=8$ ns for predistortion') #, label = '%s'%(pulse_leg))
plt.xlabel(r'Pulse duration, $t_\mathrm{p}$ (ns)', fontsize = 16)
plt.ylabel(r'Decay time, $\tau_1$ (ns)', fontsize = 16)
plt.legend(fontsize = 12, loc = 'upper right') #'upper right')
plt.title('CosRiseFall DRAG-P', fontsize=15)
plt.ylim([0, 14])
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
ax.tick_params(length=6, width = 1.4)
for side in ['top', 'left', 'right', 'bottom']:
    ax.spines[side].set_linewidth(1.4)
```
