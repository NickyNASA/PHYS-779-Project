```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import os
```

```python
#Re-read csv-files and plot
light_blue = [0.3010, 0.7450, 0.9330]
dark_blue = [51/255,51/255,255/255] #[65,105,225]/255; %royal blue



fig, ax = plt.subplots(1, 1, figsize=(5.2,4))

for idx in range(3):
    #Experiments
    axis_shift_vs_delay_arr = np.genfromtxt(os.path.join('idx_of_a1_%d'%(idx), 'axis_shift_vs_delay_arr.csv'), delimiter=',')
    delay_arr = np.genfromtxt(os.path.join('idx_of_a1_%d'%(idx), 'delay_arr.csv'), delimiter=',')
    a1_value = np.genfromtxt(os.path.join('idx_of_a1_%d'%(idx), 'a1_value.csv'), delimiter=',')
    plt.plot(delay_arr*1e9, axis_shift_vs_delay_arr, 'o--', color=np.array(light_blue)*idx/2 + np.array(dark_blue)*(2-idx)/2, label=r'$a_1 = %.2f$'%(a1_value))


plt.xlabel(r'Delay between pulses, $t_\mathrm{d}$ (ns)', fontsize = 16)
plt.ylabel(r'Axis shift, $\varphi_\mathrm{s}$ (deg.)', fontsize = 16)
plt.legend(fontsize = 11.5, loc = 'lower right') #'upper right')
#plt.title('Duration without delay: %.2f ns'%(tg_without_pad_no_pred*1e9), fontsize=14)
plt.ylim([-1.1, 1.1])
plt.title('FAST DRAG-P, $t_\mathrm{p}=9.6$ ns', fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
ax.tick_params(length=6, width = 1.4)
for side in ['top', 'left', 'right', 'bottom']:
    ax.spines[side].set_linewidth(1.4)
```
