```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import os
```

```python
e_state_probability_vs_Nr_and_td_mat = np.genfromtxt('e_state_probability_vs_Nr_and_td_mat.csv', delimiter=',')
number_of_repeats_arr = np.genfromtxt('number_of_repeats_arr.csv', delimiter=',')
delay_arr = np.genfromtxt('delay_arr.csv', delimiter=',')
pulse_duration = np.genfromtxt('pulse_duration.csv', delimiter=',')


fig, ax = plt.subplots(1, 1, figsize=(5,4))
X, Y =  np.meshgrid(delay_arr*1e9, number_of_repeats_arr)
pcmesh = plt.pcolormesh(X, Y, e_state_probability_vs_Nr_and_td_mat, shading = 'auto', rasterized=True)
#plt.title('%s, %s, last run: %s' % (qubit, chip_label, str(spec_run_id)), fontsize = 9)
plt.xlabel('Delay between pulses, $t_\mathrm{d}$ (ns)', fontsize = 16)
plt.ylabel('Number of repeats, $N_\mathrm{r}$', fontsize = 16)
cbar = fig.colorbar(pcmesh)
cbar.solids.set_rasterized(True)
cbar.set_label(label = 'Excited state probability, $p_\mathrm{e}$', size=16)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.set_cmap("coolwarm") #bwr, coolwarm 
plt.title('Pulse duration: $t_\mathrm{p} = %.2f$ ns'%(pulse_duration*1e9), fontsize=15)
ax.tick_params(length=6, width = 1.4)
for side in ['top', 'left', 'right', 'bottom']:
    ax.spines[side].set_linewidth(1.4)



```
