```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
#Re-read csv-files and plot
pulse_names = ['FAST_DRAG-L', 'FAST_DRAG-P', 'FAST_DRAG-L_no_predistortion', 'FAST_DRAG-P_no_predistortion']

light_blue = [0.3010, 0.7450, 0.9330]
dark_blue = [51/255,51/255,255/255] #[65,105,225]/255; %royal blue
crimson = [220/255, 20/255, 60/255] 
light_red = [240/255,128/255,128/255]

colors_dict = {'FAST_DRAG-L': dark_blue,
               'FAST_DRAG-P': light_blue,
               'FAST_DRAG-L_no_predistortion':crimson,
               'FAST_DRAG-P_no_predistortion': light_red,

    }


fig, ax = plt.subplots(1, 1, figsize=(5.2,4))
for pulse in pulse_names:
    #Experiments
    axis_shift_arr = np.genfromtxt('axis_shift_%s_arr.csv'%(pulse), delimiter=',')
    axis_shift_uncertainty_arr = np.genfromtxt('axis_shift_uncertainty_%s_arr.csv'%(pulse), delimiter=',')
    duration_arr = np.genfromtxt('duration_%s_arr.csv'%(pulse), delimiter=',')
    plt.errorbar(duration_arr*1e9, np.abs(axis_shift_arr), yerr=axis_shift_uncertainty_arr, fmt='o--', color=colors_dict[pulse], label = '%s'%(pulse))

    
plt.xlabel('Gate duration, $t_\mathrm{g}$ (ns)', fontsize = 16)
plt.ylabel(r'Axis shift, $\varphi_\mathrm{s}$ (deg.)', fontsize = 16)
plt.legend(fontsize = 12, loc = 'upper right') #'upper right')
ax.tick_params(length=6, width = 1.4, labelsize=13)
ax.tick_params(length=6, width = 1.4)
for side in ['top', 'left', 'right', 'bottom']:
    ax.spines[side].set_linewidth(1.4)
plt.xlim([5, 20])
plt.ylim([0, 2.5])
```
